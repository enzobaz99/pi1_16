Cargamos y transformamos los datos de reviews solo que con un ligero cambio al ver que los datos estan expresados de otra manera

```python
#Abrimos el archivo "australian_user_reviews.json" en modo lectura ('r') con codificaciÃ³n UTF-8 y lo asigna a la variable w.
#El contexto with se utiliza para garantizar que el archivo se cierre adecuadamente despuÃ©s de su uso.
with open("australian_user_reviews.json", 'r', encoding='utf-8') as w:
    data = w.readlines()
    
datos = [eval(line.strip()) for line in data]

df_review = pd.DataFrame(datos)
df_review.head()
```

Al ver que el archivo tiene nudos en otras palabras que no se pudo ver como esperaba desanidamos

```python
#creo un df con el archivo review y accedo a la columna que me interesa desanidar.
df_reviewsOpen = df_review.explode('reviews')
#con esta expresion creo un nuevo df con la info desanidada en columnas
dfreviewsOpen = pd.concat([df_reviewsOpen.drop(['reviews'], axis=1), df_reviewsOpen['reviews'].apply(pd.Series)], axis=1) 
```

Al ver que la columna posted esta en un fomrato que es dificil de trabajar prefiero pasarlo a formato datatime y al ver que para el
pedido que me hicieron solo necesitare aÃ±o dejo la columna posted como posted year 

```python
# Creamos la columna posted year
dfreviewsOpen['posted year'] = dfreviewsOpen['posted'].str.extract(r'(\d{4})')

# Elimino la columna posted
dfreviewsOpen.drop('posted' , axis = 1, inplace = True)
```

Cargamos y transformamos items

```python
with open("australian_users_items.json", 'r', encoding='utf-8') as i:
    data = i.readlines()
    
records = [eval(line.strip()) for line in data]

df_item = pd.DataFrame(records)
df_item.head()
```

Desanidamos items

```python
# Aplico la funcion lambda
df_item['playtime_forever'] = df_item['items'].apply(lambda x: x[0].get('playtime_forever') if len(x) > 0 else 0)
df_item['playtime_2weeks'] = df_item['items'].apply(lambda x: x[0].get('playtime_2weeks') if len(x) > 0 else 0)
df_item['item_name'] = df_item['items'].apply(lambda x: x[0].get('item_name') if len(x) > 0 else 0)
df_item['item_id'] = df_item['items'].apply(lambda x: x[0].get('item_id') if len(x) > 0 else 0)
df_item['playtime_forever'] = df_item['playtime_forever'].astype(int)
df_item['playtime_2weeks'] = df_item['playtime_2weeks'].astype(int)
```

```python
df_item.drop(['items'], axis=1, inplace=True)
```

Unimos df_item y df_games

```python
merged_df = pd.merge(df_item, df_games, left_on='item_name', right_on='app_name', how='inner')
```

```python
df_completo = pd.merge(merged_df, dfreviewsOpen, on='user_id', how='inner')
```

```python
df_completo.drop(['user_url_y'], axis=1, inplace=True)
df_completo.drop(['early_access'], axis=1, inplace=True)
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Initialize the sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Function to assign values according to the scale
def get_sentiment_score(text):
    if pd.isnull(text) or text == '':
        return 1  # Return neutral if it is empty or NaN
    elif isinstance(text, str):
        sentiment = sia.polarity_scores(text)
        compound_score = sentiment['compound']
        if compound_score >= -0.05:
            return 0  # Good score
        elif compound_score <= -0.05:
            return 2  # Bad score
        else:
            return 1
    else:
        return 1  # Return neutral for non-string values


# Convertir la columna 'review' a tipo de dato str
df_completo['review'] = df_completo['review'].astype(str)

# Aplicar la funciÃ³n get_sentiment_score a la columna 'review'
df_completo['sentiment_score'] = df_completo['review'].apply(get_sentiment_score)

# Imprimir el DataFrame actualizado
df_completo
```

```python
df_completo.drop(['review'], axis=1, inplace=True)
df_completo.drop(['item_id_x'], axis=1, inplace=True)
df_completo.drop(['title'], axis=1, inplace=True)
df_completo.drop(['items_count'], axis=1, inplace=True)
df_completo.drop(['steam_id'], axis=1, inplace=True)
df_completo.drop(['user_url_x'], axis=1, inplace=True)
df_completo.drop(['playtime_2weeks'], axis=1, inplace=True)
df_completo.drop(['item_name'], axis=1, inplace=True)
df_completo.drop(['publisher'], axis=1, inplace=True)
df_completo.drop(['url'], axis=1, inplace=True)
df_completo.drop(['reviews_url'], axis=1, inplace=True)
df_completo.drop(['specs'], axis=1, inplace=True)
df_completo.drop(['funny'], axis=1, inplace=True)
df_completo.drop(['last_edited'], axis=1, inplace=True)
df_completo.drop(['helpful'], axis=1, inplace=True)
```
